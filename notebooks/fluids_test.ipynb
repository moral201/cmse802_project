# Rain Scene Simulation: Visualization

This notebook visualizes the fluid dynamics. A basic 2D particle-based fluid simulation using gravity and simple bounce collisions with the ground. This serves as a first step toward Smoothed Particle Hydrodynamics (SPH).

```python
# Fluid Simulation Visualization Notebook

import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import sys

# Add path to your physics module
sys.path.append("C:/Users/Admin/Documents/Michigan MSU/Spring 2025/CMSE 802/Project/cmse802_project/src/fluids")

from fluid_simulation import simulate_fluid

# Run the simulation once
x_positions, y_trajectory, dt = simulate_fluid()

# Set up figure
fig, ax = plt.subplots(figsize=(5, 5))
ax.set_xlim(-1, 1)
ax.set_ylim(0, 2)
ax.set_title("Preliminary 2D Fluid Simulation")

# Scatter plot for particles
scat = ax.scatter([], [], color='blue', s=10)

def init():
    y_init = y_trajectory[0]
    scat.set_offsets(np.c_[x_positions, y_init])
    return scat,


def update(frame):
    y = y_trajectory[frame]
    scat.set_offsets(np.c_[x_positions, y])
    return scat,

ani = animation.FuncAnimation(
    fig, update, frames=len(y_trajectory), init_func=init,
    blit=True, interval=dt * 1000
)

plt.tight_layout()
plt.show()

```
